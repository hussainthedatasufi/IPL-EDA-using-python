#Number of matches played in each Season

```python
plt.figure(figsize=(15,7))
sns.countplot(x='Season',data=df_matches,order=df_matches['Season'].value_counts().index.sort_values())
plt.xlabel('Season',fontsize=12)
plt.ylabel('No.of matches',fontsize=12)
plt.title('Number of matches played in each IPL season',fontsize=16)
```

#Matches won by the Teams 

```python
plt.figure(figsize=(15,7))
sns.countplot(y='winner',data=df_matches,order=df_matches['winner'].value_counts().index.sort_values(),palette='magma')
plt.xlabel('Winner',fontsize=12)
plt.ylabel('No.of matches',fontsize=12)
plt.title('Number of matches won by the Teams',fontsize=16)
```

#No of matches played in stadium, Top 10

```python
plt.figure(figsize=(12,8))
sns.countplot(y='venue',data=df_matches,order=df_matches['venue'].value_counts().iloc[:10].index,palette='viridis')
plt.xlabel('Number of matches')
plt.ylabel('Venues')
plt.title('Top 10 matches by venue')
```

#Toss Decision

```python
plt.figure(figsize=(6,6))
sizes=df_matches['toss_decision'].value_counts()
label=df_matches['toss_decision'].value_counts().index
plt.pie(sizes,colors=['darkcyan','salmon'],labels=label,autopct='%1.1f%%',startangle=90,pctdistance=0.75,explode=(0.025,0.025))
centre_circle=plt.Circle((0,0),0.6,fc='white')
fig=plt.gcf()
fig.gca().add_artist(centre_circle)
plt.show()
```

#Toss decision to bat and field in each IPL season

```python
plt.figure(figsize=(15,7))
sns.countplot(x='Season',hue='toss_decision',data=df_matches,order=df_matches['Season'].value_counts().index.sort_values(),palette='rocket')
plt.xlabel('IPL Seasons')
plt.ylabel('count')
plt.title('Decision to field or bat in each IPL season', fontsize=16)
```

#Player of the Match throughout the seasons( TOP 10)

```python
plt.figure(figsize=(15,7))
player_of_match=df_matches['player_of_match'].value_counts()[:10]
sns.barplot(player_of_match.index,player_of_match.values,palette='mako')
plt.xlabel('Player',fontsize=12)
plt.ylabel('count',fontsize=12)
plt.title('Player of the Match', fontsize=16)
```
